```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
```

```python
df = pd.read_csv("../data/Tiger_20_05_00:39:26.csv")
times = df['times']
df = df.drop('times', axis=1)

new_df = pd.DataFrame(columns=['times','policy', 'result'])
for col in df.columns:
    tmp = pd.concat([times,df[col]], axis=1)
    tmp = tmp.rename(columns={col:'result'})
    tmp['policy'] = [col]*len(tmp)
    new_df = pd.concat([new_df,tmp], axis=0)

fig, ax = plt.subplots(1,1, figsize=(12,7))
sns.lineplot(data=new_df, x='times',y='result', hue='policy', ci=95, marker='o')
ax.set_xlabel("Planning Time", fontsize=15)
ax.set_ylabel("Reward", fontsize=15)
ax.set_title("Tiger Benchmark", fontsize=20)
```
