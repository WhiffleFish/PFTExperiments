```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
sns.set_context('talk')
```

```python
df0 = pd.read_csv("../data/Baby_20_05_14:10:33.csv")
times = df0['times']
df = df0.drop('times', axis=1)

new_df = pd.DataFrame(columns=['times','policy', 'result'])
for col in df.columns:
    tmp = pd.concat([times,df[col]], axis=1)
    tmp = tmp.rename(columns={col:'result'})
    tmp['policy'] = [col]*len(tmp)
    new_df = pd.concat([new_df,tmp], axis=0)

fig, ax = plt.subplots(1,1, figsize=(12,7))
sns.lineplot(data=new_df,x='times',y='result', hue='policy', ci=95)
# ax.set_xlabel("Planning Time", fontsize=15)
# ax.set_ylabel("Reward", fontsize=15)
# ax.set_title("Baby Benchmark", fontsize=20)
ax.set_xlabel("Planning Time")
ax.set_ylabel("Reward")
ax.set_title("Baby Benchmark")
```
