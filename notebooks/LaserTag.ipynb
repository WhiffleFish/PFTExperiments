```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
sns.set(context='talk')
```

```python
df = pd.read_csv("../scripts/LaserTag/data/compare_01.csv")
df.head()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,7))
# new_df = df.copy()
res = sns.lineplot(x='t',y='r',hue='sol',data=df, ax=ax, ci=95, lw=5)
res.set(xscale='log')
plt.xlabel("Planning Time (s)")
ax.legend(title="Solver Type")
plt.ylabel("Mean Reward")
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,7))
new_df = df[df['sol'] != "SparsePFT"]
res = sns.lineplot(x='t',y='r',hue='sol',data=new_df, ax=ax, ci=68, lw=5)
res.set(xscale='log')
ax.set_xlabel("Planning Time (s)")
ax.legend(title="Solver Type")
ax.set_ylabel("Mean Reward")
plt.show()
```
